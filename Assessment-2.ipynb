```python
import pandas as pd
df = pd.read_csv("Iris.csv")
df = df.drop("Id",axis = 1)
df.head()
```

```python
### show the count of different types of species
df.groupby(["Species"]).size()
```

```python
### find the max,m and mean in values of Sepallength, petallength, sepalwidth and petalwidth

## sepal length
print("mean of sepallength :",df["SepalLengthCm"].mean())
print("max of sepallength :",df["SepalLengthCm"].max())
print("min of sepallength :",df["SepalLengthCm"].min())
## sepal width
print("mean of SepalWidth :",df["SepalWidthCm"].mean())
print("max of SepalWidth :",df["SepalWidthCm"].max())
print("min of SepalWidth:",df["SepalWidthCm"].min())
## petal length
print("mean of petal length :",df["PetalLengthCm"].mean())
print("max of petal length :",df["PetalLengthCm"].max())
print("min of petal length:",df["PetalLengthCm"].min())
## petal width
print("mean of petal width :",df["PetalWidthCm"].mean())
print("max of petal width :",df["PetalWidthCm"].max())
print("min of petal width:",df["PetalWidthCm"].min())
```

```python
## scatterplot(for sepallength and petallength)
import matplotlib.pyplot as plt
x = df["SepalLengthCm"]
y = df["PetalLengthCm"]
plt.scatter(x,y)
plt.xlabel("Sepal length")
plt.ylabel("Petal length")
plt.title("Scatter plot of Sepal length and petal length")
    
```

```python
## bargraph

plt.bar(df["Species"],df["SepalLengthCm"])
plt.bar(df["Species"],df["SepalWidthCm"])
plt.bar(df["Species"],df["PetalLengthCm"])
plt.bar(df["Species"],df["PetalWidthCm"])
```

```python
## jointplot(for sepallength and sepalwidth)
import seaborn as sns
sns.jointplot(df["SepalLengthCm"],df["SepalWidthCm"])
```

```python
### facetgrid(for sepallength and sepalwidth for all species)
m = sns.FacetGrid(df,hue = "Species")
m = m.map(plt.hist,"SepalWidthCm")
m = m.add_legend()
plt.show()

v = sns.FacetGrid(df,hue = "Species")
v = v.map(plt.hist,"SepalLengthCm")
v = v.add_legend()
```

```python
### boxplot(for species and sepalwidth)

bp = sns.boxplot(df["Species"],df["SepalWidthCm"])
```

```python
### boxplot & stripplot(for sepallength and all species)

## boxplot && stripplot

bp = sns.boxplot(df["SepalLengthCm"],df["Species"])
sp = sns.stripplot(x = df["SepalLengthCm"] , y= df["Species"], data=df)
```

```python
## violinplot(for petallength and all species)

vp = sns.violinplot(x = df["Species"], y= df["PetalLengthCm"], data=df)
```

```python
### pairplot(for all 4 columns with species)

sns.pairplot(df,hue = "Species")
```
